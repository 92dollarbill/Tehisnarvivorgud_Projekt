```python
## Load needed packages
import numpy as np 
import pandas as pd
import os
import shutil
from os import path
from sklearn import model_selection
import matplotlib.pyplot as plt
import keras
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from keras.layers import GlobalAveragePooling2D, Convolution2D, Flatten, Dense, Dropout, BatchNormalization, Activation, MaxPool2D
from tensorflow.keras import layers
from PIL import Image
from keras.models import Sequential
from keras.optimizers import RMSprop, Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.applications import EfficientNetB0
from keras.callbacks import TensorBoard
import codecs
import json
from sklearn import model_selection
```

```python
# Read in data file and generate DF 
y_train_test = './data/256_images/iwildcam-2020-fgvc7/train2020.json'
y_train_test = './data/256_images/iwildcam2020_train_annotations.json'

with open(y_train_test) as file:
        y_train_test_annotations = json.load(file)
     
# Load categoreis    
categories_df = pd.DataFrame(y_train_test_annotations["annotations"])[["image_id","category_id"]]

# Load image file names
images_df = pd.DataFrame(y_train_test_annotations["images"])[["id", "file_name"]].rename(columns={"id":"image_id"})

# Merge category and image data
y_train_test_images_categories = pd.merge(categories_df, images_df, on="image_id")

# Convert categories to string
y_train_test_images_categories["category_id"]=y_train_test_images_categories["category_id"].astype(str)

# For testing
print("Training data: ")
y_train_test_images_categories
```

```python
# Lets see distribution between categories
pd.set_option('display.max_rows', None)
print("Frequency distribution in training data: ")

freq_table = pd.crosstab(index=y_train_test_images_categories["category_id"], columns="count")
freq_table = freq_table.sort_values("count")
freq_table
```

```python
# Add image present or not column
pd.set_option('display.max_rows', 200)
y_train_test_images_categories["image_present"] = np.where(y_train_test_images_categories["category_id"]== "0", "0", "1")
y_train_test_images_categories
```

```python
# Create second file where only pictures with animals present are displayed
y_train_test_images_picture_present = y_train_test_images_categories[y_train_test_images_categories["category_id"]!="0"]
y_train_test_images_picture_present
```

```python
## For training, data related parameters
BATCH_N = 64 # batch size
IMAGE_S = 224 # Image size for animal category model
IMAGE_S_PRESENT = 56 # Image size for model taht categorises wether animal is present or not

## Find number of animal categories present in the data
categories = y_train_test_images_picture_present.category_id.unique()

# Categories data for training
CATEGORIES = y_train_test_images_picture_present.category_id.unique()
print("Categories are: ")
print(CATEGORIES)

# Number of categories
CATEGORIES_N = len(CATEGORIES) # 267 categories in actual validation set, so some classes do not have training examples
print("Number of categories: ")
print(CATEGORIES_N) 
```

```python
## Image generator to classify wether the pircture has an animal present or not
df_train_animal_present, df_test_animal_present = model_selection.train_test_split(y_train_test_images_categories[:115000], test_size=0.3, random_state=44, shuffle=True)

train_datagen=ImageDataGenerator(
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    vertical_flip = True,
    rescale=1./255
    )

test_datagen = ImageDataGenerator(rescale=1./255) ## No image augmentation for test images

train_generator_animal_present=train_datagen.flow_from_dataframe(    
    dataframe= df_train_animal_present,    
    directory="./data/256_images/train/images",
    x_col="file_name",
    y_col="image_present",
    batch_size=BATCH_N,
    shuffle=False,
    class_mode="binary",
    target_size=(IMAGE_S_PRESENT,IMAGE_S_PRESENT))

valid_generator_animal_present=test_datagen.flow_from_dataframe(    
    dataframe= df_test_animal_present,    
    directory="./data/256_images/train/images",
    x_col="file_name",
    y_col="image_present",
    batch_size=BATCH_N,
    shuffle=False,
    class_mode="binary", 
    target_size=(IMAGE_S_PRESENT,IMAGE_S_PRESENT))
```

```python
## Model to classify wether the pircture has an animal present or not 
model_animal_present=Sequential()

## 1st Convolution with 32 activation fields, keeping inital input dimensions
model_animal_present.add(Convolution2D(filters=32,
                                       kernel_size=(3,3),
                                       input_shape=(IMAGE_S_PRESENT,IMAGE_S_PRESENT,3),
                                       padding='same'))

## Batch normalization
model_animal_present.add(BatchNormalization())

## ReLU activation of convolution output 
model_animal_present.add(Activation('relu'))

## Dropout layer
model_animal_present.add(Dropout(rate=0.35))

## Reduze the w and h dimensions by two by taking max in every 2 x 2 square of every activation field
model_animal_present.add(MaxPool2D(pool_size=(2,2)))

## Double the activation fields, keeping w x h dimensions
model_animal_present.add(Convolution2D(filters=64,kernel_size=(3,3),padding='same'))

## Batch normalization
model_animal_present.add(BatchNormalization())

## ReLU activation of convolution output 
model_animal_present.add(Activation('relu'))

## Dropout layer
model_animal_present.add(Dropout(rate=0.45))

## Reduze the w and h dimensions by two by taking max in every 2 x 2 square of every activation field
model_animal_present.add(MaxPool2D(pool_size=(2,2),padding='same'))

## Turn 3D tensor into 2D vector
model_animal_present.add(Flatten())

## Add dense hidden layer with 256 units
model_animal_present.add(Dense(256,activation='relu'))

## Batch normalization
model_animal_present.add(BatchNormalization())

## Add activation layer
model_animal_present.add(Activation('relu'))

## Add dropout
model_animal_present.add(Dropout(rate=0.45))

## Add dense hidden layer with 256 units
model_animal_present.add(Dense(256,activation='relu'))

## Batch normalization
model_animal_present.add(BatchNormalization())

## Add activation layer
model_animal_present.add(Activation('relu'))

## Add dropout
model_animal_present.add(Dropout(rate=0.45))

## Output layer
model_animal_present.add(Dense(1,activation='sigmoid'))

## Model sumamry and compilation
model_animal_present.summary()
model_animal_present.compile()
```

```python
TRAIN_STEP_N = train_generator_animal_present.n // train_generator_animal_present.batch_size
VALID_STEP_N = valid_generator_animal_present.n // valid_generator_animal_present.batch_size

def fit_animal_present_model(model):
       
    # Compile the model
    model.compile(
        optimizer = Adam(learning_rate = 1e-3),
        loss = 'binary_crossentropy',
        metrics = ['accuracy']
    )
    
    ## Callbacks for adapive learning rate, vizualisation and to avoid mid-training plateua in training loss
    
    # Stop training when the val_loss has stopped decreasing for 4 batches.
    early_stoppinh_callback = EarlyStopping(
        monitor='val_loss', 
        mode='min', 
        patience=4,
        restore_best_weights=True, 
        verbose=1,
    )
    
    ## Define callback for TensorBoard
    tensorboard_cbk = TensorBoard(log_dir="logs/training-example/")
    
    # Save the model with the minimum validation loss
    model_checkpoint_callback = ModelCheckpoint(
        "Best_model_animal_present.h5",
        save_best_only=True,
        monitor='val_loss',
        mode='min',
    )
    
    # Reduce learning rate once learning stagnates
    reduce_on_plateau_callback = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=2,
        min_lr=1e-6,
        mode='min',
        verbose=1,
    )
    
    # Fitting the model itself
    history = model.fit(
        train_generator_animal_present,
        validation_data=valid_generator_animal_present,
        epochs=EPOCH_N,
        batch_size=BATCH_N,
        steps_per_epoch=TRAIN_STEP_N,
        validation_steps=VALID_STEP_N,
        callbacks=[early_stoppinh_callback, model_checkpoint_callback, reduce_on_plateau_callback, tensorboard_cbk],
    )
    
    model.save('Best_model_animal_present'+'.h5')  
    
    return history
```

```python
## Fit animal present or not model 
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))

## Hack to get past image loading error
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
tf.debugging.set_log_device_placement(True)

## Training runs in epochs
EPOCH_N = 10

## Calling training wrapper fucntion
history_animal_present = fit_animal_present_model(model_animal_present)
print('Train Binary Accuracy: ', max(history_animal_present.history['accuracy']))
print('Test Binary Accuracy: ', max(history_animal_present.history['accuracy']))

## Save model history
np.save('animal_present_history.npy', history_animal_present.history)
```
